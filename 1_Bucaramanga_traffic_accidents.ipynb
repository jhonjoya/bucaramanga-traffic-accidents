Importamos las herramientas de manejo de información necesarias para manejar la información y dar respuesta a nuestras preguntas.
Nuestras librerías son:

```python
import pandas as pd
import numpy as np
import seaborn as sns
pd.__version__
```

```python
raw = pd.read_csv('db/bucaramanga_transito_accidentes.csv')
raw.head(10)
```

```python
raw.drop(['ORDEN'], axis=1 ,inplace=True)
raw.tail(5)
```

```python
raw.info()
```

```python
raw['AÑO'].unique()
```

## Corregir categorías de día

```python
raw['DÍA'].unique()
```

```python
def conv_dia(x):
    if x == "05. 05. Viernes":
        return "05. Viernes"
    else:
        return x
```

```python
raw['DÍA'] = raw['DÍA'].apply(conv_dia)
raw['DÍA'].unique()
```

### Corregir Categoría Gravedad

```python
def corr_cat_grav(x):
    return x.upper()
```

```python
raw['GRAVEDAD'] = raw['GRAVEDAD'].apply(corr_cat_grav)
raw['GRAVEDAD'].unique()
```

### Corrección de Hora y llenado de datos null

```python
def to_hour(X):
    return X[11:13]
```

```python
raw['FECHA']=pd.to_datetime(raw['FECHA'])
raw.pop('AÑO')
raw.pop('MES')
raw['HORA']=raw['HORA'].fillna(method='ffill') # LLenar el campo Null de HORA
temp=raw['HORA'].apply(to_hour)
raw.pop('HORA')
raw.insert(2,'HORA',temp)

raw['BARRIO']= raw['BARRIO'].fillna(value='No específico')
raw['ENTIDAD'] = raw['ENTIDAD'].fillna(value='No específica')
raw.head(5)
```

### Se agrupa información en Categorías correlacionadas y renombran columnas
* Se agrupan categorías Micro y buseta
* Se agrupan categorías Campero y Camioneta
* Se renombran las columnas

```python
raw['MICRO'] = raw['MICRO']+raw['BUSETA']
raw.pop('BUSETA')
raw = raw.rename(columns={'MICRO':'MICRO_BUSETA'})
raw['CAMPAERO'] = raw['CAMPAERO'] + raw['CAMIONETA']
raw.pop('CAMIONETA')
raw = raw.rename(columns={'MICRO':'MICRO_BUSETA','CAMPAERO': 'CAMPERO_CAMIONETA','nombrecomuna':'COMUNA', 'Propietario de Vehículo':'TIPO_PROPIETARIO', 'DIURNIO/NOCTURNO':'HORARIO'})
raw.info()
```

```python
df = raw
df.sample(5)
```
