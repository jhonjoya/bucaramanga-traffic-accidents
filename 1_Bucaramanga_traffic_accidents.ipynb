```python
raw = pd.read_csv('db/bucaramanga_transito_accidentes.csv')
raw.sample(10)
```

## Corregir categorías de día

```python
raw['DÍA'].unique()
```

```python
def conv_dia(x):
    if x == "05. 05. Viernes":
        return "05. Viernes"
    else:
        return x
```

```python
raw['DÍA'] = raw['DÍA'].apply(conv_dia)
```

```python
raw['DÍA'].unique()
```

## Corregir Categoría Gravedad

```python
def corr_cat_grav(x):
    return x.upper()
```

```python
raw['GRAVEDAD'] = raw['GRAVEDAD'].apply(corr_cat_grav)
raw['GRAVEDAD'].unique()
```
